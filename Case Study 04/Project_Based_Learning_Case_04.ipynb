```python
# Import Dataset
dataset = pd.read_csv("MCD_2006-01-01_to_2018-01-01.csv") #import dataset based on your directory
dataset = dataset.dropna()
dataset.dropna(inplace=True)
dataset.head()
dataset.info()
```

```python
# Checking for missing values, choose for data training and set in kind of year
#a : until year training 
#b : year for testing
# Checking for missing values
print("Missing values:")
print(dataset.isnull().sum())

# Choose for data training and set in kind of year
# a: until year training
# b: year for testing
a = '2016'
b = '2017'

# Convert 'Date' column to datetime
dataset['Date'] = pd.to_datetime(dataset['Date'])

# Splitting data into training and testing sets
training_set = dataset[dataset['Date'].dt.year <= int(a)].iloc[:, 1:2].values
test_set = dataset[dataset['Date'].dt.year == int(b)].iloc[:, 1:2].values

print("Training set shape:", training_set.shape)
print("Test set shape:", test_set.shape)
```

```python
# Scaling with minmaxscaller for data train
from sklearn.preprocessing import MinMaxScaler

# Create MinMaxScaler object
sc = MinMaxScaler(feature_range=(0,1))

# Scaling the training set
training_set_scaled = sc.fit_transform(training_set)
```

```python
#checking shape of data train
print("Shape of training set after scaling:", training_set_scaled.shape)
```

```python
# create data structure to fill how many row for input in your model at data train
X_train = []
y_train = []
for i in range(60, 2769):
    X_train.append(training_set_scaled[i-60:i, 0])  # Mengambil 60 baris sebelumnya sebagai fitur untuk setiap entri
    y_train.append(training_set_scaled[i, 0])       # Mengambil nilai saham pada baris saat ini sebagai label

# Mengubah X_train dan y_train menjadi array numpy
X_train, y_train = np.array(X_train), np.array(y_train)

# Memeriksa bentuk dari X_train dan y_train
print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
```

```python
# Reshaping X_train before modelling
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

# Checking the shape of X_train after reshaping
print("Shape of X_train after reshaping:", X_train.shape)
```

## Modelling

```python
# Sequential Modelling
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout

# Sequential Modelling
model = Sequential()

# First LSTM layer with Dropout regularisation
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))

# Second LSTM layer
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))

# Third LSTM layer
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))

# Fourth LSTM layer
model.add(LSTM(units=50))
model.add(Dropout(0.2))

# The output layer
model.add(Dense(units=1))

# Compiling
model.compile(optimizer='adam', loss='mean_squared_error')

# Train fitting for the model
model.fit(X_train, y_train, epochs=100, batch_size=32)
```
